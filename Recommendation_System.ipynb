**Building a Basic Recommendation System**



```python
# let's import the libraries
import pandas as pd
import numpy as np
```

```python
# get the data
credits=pd.read_csv('/content/tmdb_5000_credits.csv.zip')
credits.head()
```

```python
movies=pd.read_csv('/content/tmdb_5000_movies.csv.zip')
movies.head()
```

```python
credits.shape
```

```python
movies.shape
```

Let's merge this 2 datasets into one using id column.

```python
id_renamed=credits.rename(index=str,columns={'movie_id':'id'})
data_merged=movies.merge(id_renamed,on='id')
data_merged.head()
```

```python
data_merged.columns
```

```python
#let's drop some unnecessary data
final_data=data_merged.drop(columns=['homepage','title_x', 'title_y', 'status','production_countries'])
final_data.head()
```

```python
final_data.info()
```

**Using Weighted average for each movie's Average Rating**



image.png


```python
v=final_data['vote_count']
c=final_data['vote_average'].mean()
r=final_data['vote_average']
m=final_data['vote_count'].quantile(0.70)
```

```python
final_data['weighted_average']=((r*v)+(c*m))/(v+m)
```

```python
final_data.head()
```

```python
movie_sorted_ranking=final_data.sort_values('weighted_average',ascending=False)
movie_sorted_ranking[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(20)
```

From this we could see the movies sorted according to its weight average. Let us visualize them.


```python
# import the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weight_average=movie_sorted_ranking.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted_average'].head(10), y=weight_average['original_title'].head(10), data=weight_average)
plt.xlim(4, 10)
plt.title('Best Movies by average votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
```

```python
popularity=movie_sorted_ranking.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(10), y=popularity['original_title'].head(10), data=popularity)

plt.title('Most Popular by Votes', weight='bold')
plt.xlabel('Score of Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_popular_movies.png')
```

**Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
movie_scaled=scaler.fit_transform(final_data[['weighted_average','popularity']])
```

```python
movie_scaled_data=pd.DataFrame(movie_scaled,columns=['weighted_average','popularity'])
movie_scaled_data.head()
```

```python
final_data[['normalized_weight_average','normalized_popularity']]= movie_scaled_data
```

```python
final_data.head()
```

```python
final_data['score']=final_data['normalized_weight_average']*0.5+final_data['normalized_popularity']*0.5
```

```python
movie_score_ranked=final_data.sort_values(['score'],ascending=False)
movie_score_ranked[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```

```python
scored_df = final_data.sort_values('score', ascending=False)

plt.figure(figsize=(16,6))

ax = sns.barplot(x=scored_df['score'].head(10), y=scored_df['original_title'].head(10), data=scored_df, palette='deep')

#plt.xlim(3.55, 5.25)
plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')

plt.savefig('scored_movies.png')
```
